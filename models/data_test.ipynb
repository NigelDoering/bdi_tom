```python
import numpy as numpy
import pandas as pandas
import matplotlib.pyplot as plt
```

```python
data_path = "./../data/simulation_data/run_8/trajectories/all_trajectories.json"
data = pandas.read_json(data_path)
data.head()
```

```python
lengths = []
for agent_id in range(100):
    agent_str = str(agent_id).zfill(3)
    for index, row in data.iterrows():
        agent_data = row[f'agent_{agent_str}']
        path = agent_data['path']
        lengths.append(len(path))
```

```python
print(lengths)
print(len(lengths))

# Plot histogram of lengths
plt.hist(lengths, bins=20, edgecolor='black')
plt.title('Histogram of Trajectory Lengths for Agent 001')
plt.xlabel('Length of Trajectory')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np
lengths_array = np.array(lengths)
mean_length = np.mean(lengths_array)
median_length = np.median(lengths_array)
std_length = np.std(lengths_array)
min_length = np.min(lengths_array)
max_length = np.max(lengths_array)
q1_length = np.percentile(lengths_array, 25)
q3_length = np.percentile(lengths_array, 75)
iqr_length = q3_length - q1_length

print(f"Mean Length: {mean_length}")
print(f"Median Length: {median_length}")
print(f"Standard Deviation: {std_length}")
print(f"Min Length: {min_length}")
print(f"Max Length: {max_length}")
print(f"1st Quartile (Q1): {q1_length}")
print(f"3rd Quartile (Q3): {q3_length}")
print(f"Interquartile Range (IQR): {iqr_length}")

plt.boxplot(lengths_array, vert=True, patch_artist=True)
plt.title('Box Plot of Trajectory Lengths for All Agents')
plt.ylabel('Length of Trajectory')
plt.grid(True)
plt.show()  
```
